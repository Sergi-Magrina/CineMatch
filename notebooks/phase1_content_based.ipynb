```python
import pandas as pd

# Load the movie dataset
df = pd.read_csv("../data/movies_sample.csv")

# Show the first 5 rows
df.head()
```

```python
# Quick info about columns and data types
df.info()

# Count rows and columns
print("Shape:", df.shape)

# Display column names
print("Columns:", list(df.columns))
```

```python
# See unique genres
print(df["genres"].unique()[:5])

# Count how many unique titles
print("Unique titles:", df["title"].nunique())

# Show 10 random rows
df.sample(10)
```

```python
# Example: "Toy Story (1995)" â†’ "Toy Story"
df["clean_title"] = df["title"].str.replace(r"\s*\(\d{4}\)$", "", regex=True)

df[["title", "clean_title"]].head()
```

```python
# Replace the pipe '|' with spaces
df["genres_clean"] = df["genres"].fillna("").str.replace("|", " ")

# Count number of genres per movie
df["num_genres"] = df["genres"].str.split("|").str.len()

df[["clean_title", "genres_clean", "num_genres"]].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Use the cleaned genre text
tfidf = TfidfVectorizer(token_pattern=r"(?u)\b\w+\b")

# Fit the vectorizer and transform genres into numbers
X = tfidf.fit_transform(df["genres_clean"])

print("Matrix shape:", X.shape)
```
